# **Simple Calculator App using Python Functions**

```python
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed"
    return a / b

# Input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

# Operation handling
if operation == "+":
    result = add(num1, num2)
elif operation == "-":
    result = sub(num1, num2)
elif operation == "*":
    result = mul(num1, num2)
elif operation == "/":
    result = div(num1, num2)
else:
    result = "Invalid operation"

# Using built-in functions
if isinstance(result, (int, float)):
    print("Absolute value:", abs(result))
    print("Rounded value:", round(result, 2))
    print("Power (num1^num2):", pow(num1, 2))

print("Result:", result)
```
